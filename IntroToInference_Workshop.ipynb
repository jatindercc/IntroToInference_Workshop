### üìê Z-Scores: Proportions & Reverse Lookups (Example: Student Score = 85)


When we have a normal distribution (not standardized), say with mean $\mu$ and standard deviation $\sigma$, we often want to answer questions like:

> *‚ÄúWhat proportion (or percentage) of students score less than $X = 85$?‚Äù*

This is where **standardization** with Z-scores comes in.


#### 1. Standardization: Transforming to the Standard Normal

The standard normal distribution has:

- mean $0$ (denoted $\mu = 0$),
- standard deviation $1$ (denoted $\sigma = 1$).

We convert any normal variable $X$ with mean $\mu$ and standard deviation $\sigma$ into a **Z-score** via:

$$
Z = \frac{X - \mu}{\sigma}
$$

This transformation shifts and scales the variable so we can use standard normal tables or functions.


#### 2. Example: Proportion Less Than a Value

Suppose:

- $X$ = test scores,
- $\mu = 70$,
- $\sigma = 10$,
- We want to find the proportion of students scoring **less than** $X = 85$.

**Step A.** Compute the Z-score for $X = 85$:

$$
Z = \frac{85 - 70}{10} = \frac{15}{10} = 1.5
$$

**Step B.** Ask: *What is $P(X < 85)$?*  
Because $X < 85 \;\;\Longleftrightarrow\;\; Z < 1.5$.

We can look up $P(Z < 1.5)$ in a standard normal table or use a calculator / programming language (e.g. `scipy.stats.norm.cdf(1.5)`).

That value is about:

$$
P(Z < 1.5) \approx 0.9332
$$

So about **93.32%** of students are expected to score less than 85.


#### 3. Reverse Lookup: From Proportion to Value

Sometimes we want the opposite: *‚ÄúWhat test score corresponds to the 95th percentile?‚Äù* In other words:

> Find $x$ such that $P(X < x) = 0.95$.

**Step A.** In standard normal world, find $z$ such that $P(Z < z) = 0.95$.  
From tables or functions, $z \approx 1.645$ (or 1.64 depending on precision).

**Step B.** Turn that back into the original $X$ scale by solving:

$$
z = \frac{x - \mu}{\sigma} 
\;\;\Longrightarrow\;\; 
x = \mu + z \sigma
$$

With $\mu = 70, \sigma = 10$:

$$
x = 70 + (1.645)(10) = 70 + 16.45 = 86.45
$$

So the 95th percentile score is about **86.45**.

## üìä T-Score

The **t-score** is a type of standard score used when the **sample size is small** or when the **population standard deviation ($\sigma$) is unknown**.  

It works very much like the **z-score**, but with one important difference:  
- Instead of using the population standard deviation ($\sigma$), it uses the **sample standard deviation ($s$)**.  
- It also adjusts for the **sample size ($n$)**, since smaller samples add more uncertainty.


### Formula

$$
T = \frac{X - \mu}{s / \sqrt{n}}
$$

Where:  
- $T$ = t-score  
- $X$ = value of the observation  
- $\mu$ = sample mean  
- $s$ = sample standard deviation  
- $n$ = sample size  


### Why use the t-score?
- When $n$ is **large** and $\sigma$ is known ‚Üí use the **z-score**.  
- When $n$ is **small** or $\sigma$ is unknown ‚Üí use the **t-score**.  

The t-distribution is wider and has heavier tails than the normal distribution, reflecting more uncertainty with smaller samples. As $n$ grows larger, the t-distribution approaches the standard normal distribution.

# üìà Test for Normality

Before applying many statistical methods, we need to check whether the data follow a **normal distribution**.  

A **normality test** helps us decide if it is reasonable to assume the data are normally distributed.

**Why Normalize a dataset?**  
In Machine Learning, features can have very different scales (e.g., age in years vs. income in dollars).  
If left unnormalized, algorithms that rely on distance (like K-Nearest Neighbors, k-Means, or gradient descent in neural networks) can become biased toward features with larger values.  

üëâ Normalization rescales all features to the same range (commonly 0 to 1), ensuring that **each feature contributes fairly** to the model.

<br/>

### üîë Assumptions: Data Independence

When performing a test for normality, we assume that the data are **independent**.  
This means:

1. **Independent Collection:**  
   - Example: Water hardness data samples across the UK were collected by **source A**.  
   - Mortality rates across the UK were collected by **source B**.  

2. **No Pairing or Matching:**  
   - The two samples (water hardness and mortality rates) are not paired or matched.  

3. **No Dependence:**  
   - The two sets of measurements do not depend on each other.

#### üîë Why Normalize the Data?

You have two features in the `water.csv` dataset:

* **Mortality** (ranges around 1200‚Äì2000)
* **Hardness** (ranges around 5‚Äì120)

These features are on **very different scales**.
If you use them directly in **distance-based models** (like K-Nearest Neighbors) or **gradient descent‚Äìbased models** (like Logistic Regression, Neural Networks), the larger-scaled feature (**Mortality**) will dominate.

üëâ This means the model may **ignore water hardness** just because mortality values are numerically larger.

**Normalization** rescales the features so they‚Äôre comparable, ensuring **each feature contributes fairly** to the model.

#### üß™ Demonstration in Python

We‚Äôll prove this using **K-Nearest Neighbors (KNN)**, a distance-based classifier.
I‚Äôll create a simple experiment: predict whether a town is in the **North or South** based on `mortality` and `hardness`.

#### 1. Without Normalization

`````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler

# Features and labels
X = water[['mortality', 'hardness']]
y = water['location']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

# KNN without normalization
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("Accuracy WITHOUT normalization:", accuracy_score(y_test, y_pred))

`````

#### 2. With Normalization (Min-Max Scaling)

`````python
# Normalize features to [0,1] range
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train/test split again (on scaled features)
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_scaled, y, random_state=42, test_size=0.3)

# KNN with normalization
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_s, y_train_s)
y_pred_s = knn.predict(X_test_s)

print("Accuracy WITH normalization:", accuracy_score(y_test_s, y_pred_s))
`````

#### üéØ What to Observe

* The **accuracy without normalization** may be poor because the large scale of `mortality` overwhelms `hardness`.
* After **normalization**, `hardness` is on equal footing with `mortality`, so the classifier can use **both features effectively**.

üí° **Takeaway:**
Always check the **scale of your features**. If they differ significantly, normalize (or standardize) before running experiments. Otherwise, you risk biased models.

### üß™ Shapiro-Wilk Test: Beyond Visual Inspection

When we look at histograms or density plots, we can often *guess* whether the data look approximately normal.  
But in **machine learning workflows**, relying on **visual inspection** is not practical:

- It is subjective ‚Äî two people may interpret the same plot differently.  
- It doesn‚Äôt scale ‚Äî imagine inspecting thousands of features across hundreds of datasets.  


#### üîë Enter the Shapiro-Wilk Test
The **Shapiro-Wilk test** provides a **formal statistical test for normality**.

- **Null hypothesis ($H_0$):** The data come from a normal distribution.  
- **Alternative hypothesis ($H_1$):** The data do not come from a normal distribution.  

It produces a **p-value**:
- If `p > 0.05`: fail to reject $H_0$ ‚Üí data are likely normal.  
- If `p <= 0.05`: reject $H_0$ ‚Üí data is **unlikely** to be normal.  

Since the p-value is smaller than the commonly used **significance level** of $\alpha = 0.05$, we reject the null hypothesis $H_0$.


#### üß† Reflection Activity: Understanding the p-value

**What is a p-value anyway?**

üëâ A p-value tells us how surprising our data would be if nothing unusual was going on (i.e., if the null hypothesis were true).

- A **small p-value** ($p \leq \alpha$) means:  
  ‚ÄúWow, this result would be very unlikely if nothing unusual was happening.‚Äù
   <br/>or more specifically:<br/>
  "Wow, this data would be very unlikely if it were truly normal."
  
- A **large p-value** ($p > \alpha$) means:  
  ‚ÄúThis result isn‚Äôt surprising at all; it could easily happen just by chance.‚Äù
   <br/>or more specifically:<br/>
  "This data isn‚Äôt surprising at all; it could easily come from a normal distribution."

### üé≤ Understanding the p-value

üëâ **What is a p-value?**

1. **Simplest view:**  
   A p-value is a number that tells you how much your result looks like it could be a **random accident**.  
   - Small p-value ‚Üí unlikely to be an accident.  
   - Large p-value ‚Üí could easily be an accident.  

2. **Everyday analogy (coin flips):**  
   Imagine flipping a fair coin. If you get **10 heads in a row**, the p-value answers:  
   *‚ÄúIf the coin were really fair, how likely would it be to see this result?‚Äù*  

3. **Plain English version:**  
   The p-value is the chance of seeing results **at least as extreme as yours** if nothing unusual is happening.  

4. **Statistical phrasing (when we can‚Äôt avoid it):**  
   The p-value is the probability of observing your data, or something more extreme, **assuming the null hypothesis is true**.  

üí° **Takeaway:**  
The smaller the p-value, the stronger the evidence that your result is *not* just random chance.


### üîó Connecting p-values and the Shapiro-Wilk Test

The **Shapiro-Wilk test** is a statistical test for checking whether a dataset is normally distributed.  
It produces a **p-value**, which we interpret just like any other p-value:

- **Null hypothesis ($H_0$):** The data come from a normal distribution.  
- **Alternative hypothesis ($H_1$):** The data do not come from a normal distribution.  

üëâ The p-value tells us how much the data could look like a **random accident** under the assumption of normality.  

- If **p > 0.05** ‚Üí the data do not provide strong evidence against $H_0$, so it is reasonable to assume the data are normal.  
- If **p ‚â§ 0.05** ‚Üí the data are unlikely under $H_0$, so we conclude the data are not normally distributed.  

üí° In short: the **Shapiro-Wilk test uses the p-value to automate the decision** of whether a dataset is ‚Äúnormal enough‚Äù to justify using methods that assume normality.


`````python
# Shapiro-Wilk test for normality
shapiro_hardness = stats.shapiro(water['hardness'])
shapiro_mortality = stats.shapiro(water['mortality'])
print(f"Shapiro-Wilk test for hardness: {shapiro_hardness}")
print(f"Shapiro-Wilk test for mortality: {shapiro_mortality}")

# Q-Q plots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Q-Q plot for hardness
stats.probplot(water['hardness'], dist="norm", plot=axes[0])
axes[0].set_title('Q-Q Plot for Hardness')

# Q-Q plot for mortality
stats.probplot(water['mortality'], dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot for Mortality')

plt.tight_layout()
plt.show()
`````

### ‚öôÔ∏è Why It Matters in ML Operations
- In automated ML pipelines, we cannot visually inspect every feature.  
- Tests like **Shapiro-Wilk** give us a **systematic, programmatic way** to decide whether normality assumptions hold.  
- This helps us choose the right tools:
  - If normal ‚Üí use **parametric methods** (e.g., linear regression, t-tests).  
  - If not normal ‚Üí consider **non-parametric methods** (e.g., Mann-Whitney test, tree-based models).  


üëâ In short: **visual inspection is good for learning**, but **automated testing is essential for scaling machine learning operations.**

## üî¨ Expanding Beyond the Workflow

You have now completed the **9-step workflow**, learning how to explore, test, transform, and reflect on data. ## üßÆ Understanding the T-Test

The **T-Test** is a statistical test used to determine whether the means of **two groups** are significantly different from each other.  
It is often applied when we have small sample sizes and we don‚Äôt know the population variance.  

In practice, a t-test helps answer questions like:  
- *‚ÄúDo two groups of students score differently on the same exam?‚Äù*  
- *‚ÄúIs there a meaningful difference between two treatment groups in an experiment?‚Äù*

### üîë Assumptions of the T-Test

For the t-test results to be valid, the following assumptions should hold:  

1. **Independence** ‚Üí The two data sets (samples) are independent of each other.  
2. **Normality** ‚Üí Each data set is approximately normally distributed.  
3. **Equal Variance (Homogeneity of Variance)** ‚Üí Both groups have similar variance.  
4. **Equal Sample Size** ‚Üí Both groups have the same number of samples.  
5. **Low Cardinality** ‚Üí Works best with relatively small sample sizes (around 20‚Äì30, not much larger).  

### üß© Why It Matters

If these assumptions are satisfied, the t-test gives a reliable way to check if an observed difference between two sample means is likely due to chance ‚Äî or if it reflects a **real underlying difference** between the groups.


### üß™ T-Test in Python

We can run a **two-sample t-test** in Python using `scipy.stats.ttest_ind()`.  
This function checks whether the means of two independent samples are significantly different.

### üíª Python Scaffold

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset
water = pd.read_csv("./data/water.csv")

# Example: Compare mortality between two groups of towns
# Split into two samples (e.g., low hardness vs. high hardness)
group_low = water[water['hardness'] < water['hardness'].median()]['mortality']
group_high = water[water['hardness'] >= water['hardness'].median()]['mortality']

# Run independent two-sample t-test
t_stat, p_val = ttest_ind(group_low, group_high, equal_var=True)

print("T-statistic:", t_stat)
print("p-value:", p_val)

# Interpretation
if p_val < 0.05:
    print("Result: Significant difference between the two groups (reject H0).")
else:
    print("Result: No significant difference (fail to reject H0).")

## üîé Understanding the F-Test

The **F-Test** is a statistical test used to compare **variances** between groups or models.  
It answers the question:  
üëâ *‚ÄúDo these groups (or models) show different amounts of variability, or are they roughly the same?‚Äù*

---

### üéØ Main Purpose

- **Comparing Variances:** The F-test checks if two or more groups have significantly different variability.  
- **ANOVA (Analysis of Variance):** The F-test is the backbone of ANOVA, which tests whether the **means of three or more groups** are significantly different.  
- **Model Comparison:** In regression and machine learning, the F-test can help decide if adding extra predictors significantly improves the model.

---

### üìä When to Use an F-Test

- When comparing **two datasets** to see if their variability is similar or different.  
- When analyzing **three or more groups** (via ANOVA) to check if their means differ.  
- When **evaluating models** to see if one explains more variance in the outcome than another.  

---

### ‚öñÔ∏è Key Assumptions

1. **Independence:** Samples from each group must be independent.  
2. **Normality:** Data in each group should follow a normal distribution.  
3. **Homogeneity of Variance:** The F-test assumes variances are similar ‚Äî though this is exactly what it‚Äôs testing.  
   *(If this assumption is violated, results may not be reliable.)*  

üëâ Normalization or scaling is **not required** for the F-test. What matters most is that the underlying data follow a normal distribution.

---

### üß© Why It Matters

The F-test provides a way to move beyond just comparing averages.  
It tells us whether the **spread of values** across groups is similar or not.  
This is critical in real-world analysis:  
- Are two processes equally stable?  
- Does one machine learning model explain more variance than another?  
- Do groups differ in their variability, not just their mean?

In short: the F-test helps us **measure and compare variability**, making it a powerful tool for inference and model evaluation.


### üß™ F-Test in Python

The F-test can be used in two common ways:  

1. **Variance Ratio Test (two groups):**  
   Compare whether the variability of two datasets is significantly different.  

2. **ANOVA (three or more groups):**  
   Test whether group means are significantly different, using the F-statistic as the decision measure.  

---

### üíª Python Scaffold

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load dataset
water = pd.read_csv("./data/water.csv")

# Example: Split mortality into two groups based on water hardness
group_low = water[water['hardness'] < water['hardness'].median()]['mortality']
group_high = water[water['hardness'] >= water['hardness'].median()]['mortality']

# --- 1. Variance Ratio Test (two groups) ---
# Compute F statistic as ratio of variances
var1 = np.var(group_low, ddof=1)
var2 = np.var(group_high, ddof=1)

F = var1 / var2
df1 = len(group_low) - 1
df2 = len(group_high) - 1

# Get p-value from F-distribution
p_val_var = 1 - stats.f.cdf(F, df1, df2)

print("Variance Ratio Test")
print("F-statistic:", F)
print("p-value:", p_val_var)

# --- 2. ANOVA (three or more groups) ---
# Split into quartiles of hardness and compare mortality means
water['quartile'] = pd.qcut(water['hardness'], 4, labels=False)
groups = [water[water['quartile'] == q]['mortality'] for q in range(4)]

f_stat, p_val_anova = stats.f_oneway(*groups)

print("\nANOVA Test")
print("F-statistic:", f_stat)
print("p-value:", p_val_anova)

## üîÑ The Wilcoxon Signed-Rank Test

The **Wilcoxon Signed-Rank Test** is a **non-parametric** test used to compare **two related samples** or repeated measurements on the same subjects.  
It answers the question:  
üëâ *‚ÄúHas something changed between two related measurements when we cannot assume normality?‚Äù*  

It is often used as an alternative to the **paired t-test** when the assumption of normality is not met.

---

### üéØ Main Purpose

- To test whether there is a **significant difference** in the **median ranks** of two related samples.  
- Useful when analyzing **before-and-after measurements**, matched pairs, or repeated measures.  
- Provides a robust way to compare paired data without requiring normally distributed values.  

---

### üìä When to Use

- **Paired data:** Two measurements from the same subjects (e.g., before vs after treatment).  
- **Non-normal data:** When the differences between pairs are not normally distributed.  
- **Ordinal or ranked data:** Works even when data are ranks rather than continuous measurements.  

---

### ‚öñÔ∏è Key Assumptions

1. The data are **paired** (each observation in one group is linked to an observation in the other).  
2. The differences between pairs are **independent** of each other.  
3. The data are at least **ordinal** (can be ranked).  

---

### üß© Why It Matters

The Wilcoxon Signed-Rank Test lets us test for meaningful differences in **paired data** even when the dataset is small, skewed, or ordinal.  
It‚Äôs especially powerful in real-world cases where normality cannot be assumed ‚Äî for example:  

- Testing the effect of a training program on student performance (before vs after scores).  
- Measuring health outcomes before and after treatment in the same patients.  
- Comparing rankings of preferences under two different conditions.  

üëâ In short: the Wilcoxon Signed-Rank Test gives us a reliable tool when the **paired t-test** is not appropriate.


### üß™ Wilcoxon Signed-Rank Test in Python

The **Wilcoxon Signed-Rank Test** checks whether the **median difference** between paired measurements is zero.  
It is most useful for **before-and-after data** that may not be normally distributed.  


### üíª Python Scaffold

```python
import numpy as np
from scipy.stats import wilcoxon

# Example: Before-and-after test scores for the same group of students
before = np.array([55, 60, 62, 70, 68, 72, 65, 58, 61, 64])
after  = np.array([60, 63, 65, 74, 70, 75, 67, 62, 64, 66])

# Run the Wilcoxon Signed-Rank Test
stat, p_val = wilcoxon(before, after)

print("Wilcoxon statistic:", stat)
print("p-value:", p_val)

# Interpretation
if p_val < 0.05:
    print("Result: Significant difference between 'before' and 'after' (reject H0).")
else:
    print("Result: No significant difference detected (fail to reject H0).")

## üî¢ The Chi-Squared Test

The **Chi-Squared Test (œá¬≤ test)** is a statistical method used to test whether there is a **significant association** between two categorical variables.  
It works by comparing what we **observed** in our data to what we would **expect** if there were no relationship between the variables.  

### üéØ Main Purpose

- To determine whether differences between observed and expected frequencies are simply due to **random chance** or whether they reveal an **underlying relationship**.  
- Commonly used for:
  - **Independence tests:** Are two variables associated?  
  - **Goodness-of-fit tests:** Does the observed data follow a given distribution?  

### üìä When to Use

- When you have **categorical data** (e.g., gender, region, product category).  
- To test if two variables are **independent** or associated.  
- To compare an **observed distribution** with an **expected distribution** under a hypothesis.  

### ‚öñÔ∏è Key Assumptions

1. Data are **frequencies** (counts), not percentages or continuous values.  
2. Observations are **independent** of each other.  
3. Expected frequencies in each cell should generally be **5 or more** to ensure validity.  

### üß© Why It Matters

The Chi-Squared test helps us answer questions like:  
- *‚ÄúAre customer preferences independent of product type?‚Äù*  
- *‚ÄúIs there a relationship between region and voting behavior?‚Äù*  
- *‚ÄúDoes my dataset fit the distribution I expected?‚Äù*  

üëâ In short: the Chi-Squared test provides a powerful tool for detecting relationships in **categorical data** that cannot be tested with parametric methods like t-tests or ANOVA.


### üß™ Chi-Squared Test in Python

The **Chi-Squared Test (œá¬≤ test)** can be applied to a **contingency table** (a table of frequencies).  
It checks whether two categorical variables are **independent** or **associated**.

### üíª Python Scaffold

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Example dataset: Survey of preferences by gender
# Rows = Gender (Male, Female)
# Columns = Preference (A, B, C)
data = np.array([
    [30, 10, 20],   # Male counts
    [25, 20, 15]    # Female counts
])

# Convert to DataFrame for readability
df = pd.DataFrame(data, index=["Male", "Female"], columns=["A", "B", "C"])
print("Contingency Table:")
print(df)

# Run Chi-Squared Test of Independence
chi2, p_val, dof, expected = chi2_contingency(df)

print("\nChi-Squared Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("p-value:", p_val)
print("\nExpected Frequencies (if independent):")
print(expected)

# Interpretation
if p_val < 0.05:
    print("Result: Significant association between gender and preference (reject H0).")
else:
    print("Result: No significant association (fail to reject H0).")

## üîó The Connection Between the T-Test and the Levene Test

The **T-test** is used to compare the means of two groups to see if they are significantly different.  
But the T-test has an important **assumption**: both groups should have **similar variances** (homogeneity of variance).  

This is where the **Levene Test** comes in.  

### üéØ Levene Test

- The **Levene Test** checks whether the variances of two (or more) groups are equal.  
- Null hypothesis ($H_0$): The groups have equal variances.  
- If the p-value is **high (> 0.05)** ‚Üí assume equal variances (use the standard T-test).  
- If the p-value is **low (‚â§ 0.05)** ‚Üí variances differ (use the *Welch‚Äôs T-test*, a version of the T-test that does not assume equal variances).  

### üß© Why They Belong Together

- The T-test asks: *‚ÄúAre the means of these groups different?‚Äù*  
- The Levene Test asks: *‚ÄúDo these groups have similar spreads (variances)?‚Äù*  

üëâ Together, they ensure we apply the **right version of the T-test**:  
- **Equal variances ‚Üí Standard T-test**  
- **Unequal variances ‚Üí Welch‚Äôs T-test**  

### üìå Key Insight

The Levene Test is not about comparing means directly, but about checking an **assumption** that affects the reliability of the T-test.  
By running Levene first, you make sure the T-test results are valid.


`````python
# Variance test
res_ftest = stats.levene(water['mortality'][water['location'] == 'North'],
                         water['mortality'][water['location'] == 'South'])
print(f"Variance test result: {res_ftest}")

# T-Test
res_ttest = stats.ttest_ind(water['mortality'][water['location'] == 'North'],
                            water['mortality'][water['location'] == 'South'],
                            equal_var=True)
print(f"T-Test result: {res_ttest}")
`````

## üìä Summary of Statistical Tests

| Test                        | When to Use                                                                 | Key Assumptions                                                                 |
|-----------------------------|------------------------------------------------------------------------------|---------------------------------------------------------------------------------|
| **T-Test**                  | Compare the **means** of two independent groups.                             | - Independent samples<br>- Normally distributed data<br>- Equal variances (unless Welch‚Äôs T-test)<br>- Sample sizes similar |
| **F-Test**                  | Compare the **variances** of two groups, or test variance explained by models.| - Independent samples<br>- Normally distributed data<br>- Homogeneity of variance |
| **ANOVA**                   | Compare the **means** of **three or more groups**.                          | - Independent samples<br>- Normally distributed data in each group<br>- Homogeneity of variance |
| **Wilcoxon Signed-Rank**    | Compare **paired or related samples** (before vs after, matched pairs).      | - Paired/related samples<br>- Differences are independent<br>- Data at least ordinal<br>- Normality **not required** |
| **Chi-Squared Test**        | Test association between **two categorical variables**, or goodness-of-fit.  | - Data are counts/frequencies<br>- Observations are independent<br>- Expected cell frequencies ‚â• 5 |
| **Levene Test**             | Test if **variances are equal** across two or more groups (pre-check for T-test/ANOVA). | - Independent samples<br>- Data roughly symmetric<br>- Less sensitive to non-normality than F-test |
